```python
def Fibonacci(N:int):
  if N < 2:
    return 1
  else:
    return Fibonacci(N-1)+Fibonacci(N-2)

Fibonacci(5)
```

# Devolución de cambio por técnica voraz

```python
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(15,[25,10,5,1])
```

# N-Reinas por técnica de vueta atrás


```python
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        #escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(8)
```

# 1. Algoritmo voraz: Mochila fraccionaria

## 1.1 Descripción del problema

El problema de la **mochila fraccionaria** consiste en seleccionar objetos con diferentes pesos y valores para maximizar el valor total transportado, sin superar una capacidad máxima de peso.

A diferencia de la mochila clásica 0/1, en este caso se permite tomar una **fracción** de un objeto. Por eso, una estrategia voraz funciona correctamente.

## 1.2 Idea del algoritmo

La estrategia consiste en calcular la relación:

$$
\text{valor por peso} = \frac{\text{valor}}{\text{peso}}
$$

Luego se ordenan los objetos de mayor a menor según esa relación y se van agregando a la mochila mientras haya capacidad disponible.

Este algoritmo es **voraz** porque en cada paso toma la mejor decisión local: elegir el objeto con mayor valor por unidad de peso.


```python
def mochila_fraccionaria(objetos, capacidad):
    """
    Resuelve el problema de la mochila fraccionaria usando una estrategia voraz.

    Parámetros:
    objetos: lista de diccionarios con nombre, valor y peso.
    capacidad: capacidad máxima de la mochila.

    Retorna:
    valor_total: valor máximo obtenido.
    seleccionados: lista de objetos seleccionados, indicando si se tomó completo o fraccionado.
    """

    # Calculamos la relación valor/peso para cada objeto
    for objeto in objetos:
        objeto["valor_por_peso"] = objeto["valor"] / objeto["peso"]

    # Ordenamos los objetos de mayor a menor valor por peso
    objetos_ordenados = sorted(
        objetos,
        key=lambda x: x["valor_por_peso"],
        reverse=True
    )

    valor_total = 0
    capacidad_restante = capacidad
    seleccionados = []

    for objeto in objetos_ordenados:
        if capacidad_restante == 0:
            break

        # Si el objeto entra completo, se toma completo
        if objeto["peso"] <= capacidad_restante:
            seleccionados.append({
                "objeto": objeto["nombre"],
                "fraccion": 1,
                "peso_tomado": objeto["peso"],
                "valor_obtenido": objeto["valor"]
            })

            valor_total += objeto["valor"]
            capacidad_restante -= objeto["peso"]

        # Si no entra completo, se toma solo una fracción
        else:
            fraccion = capacidad_restante / objeto["peso"]
            valor_obtenido = objeto["valor"] * fraccion

            seleccionados.append({
                "objeto": objeto["nombre"],
                "fraccion": fraccion,
                "peso_tomado": capacidad_restante,
                "valor_obtenido": valor_obtenido
            })

            valor_total += valor_obtenido
            capacidad_restante = 0

    return valor_total, seleccionados


# Datos de prueba
objetos = [
    {"nombre": "A", "valor": 60, "peso": 10},
    {"nombre": "B", "valor": 100, "peso": 20},
    {"nombre": "C", "valor": 120, "peso": 30}
]

capacidad = 50

valor_total, seleccionados = mochila_fraccionaria(objetos, capacidad)

print("Objetos seleccionados:")
for item in seleccionados:
    print(item)

print("\nValor total obtenido:", valor_total)
```

## 1.3 Análisis

El algoritmo selecciona primero los objetos con mayor relación valor/peso.  
En el ejemplo, los objetos se ordenan así:

1. Objeto A: 60 / 10 = 6
2. Objeto B: 100 / 20 = 5
3. Objeto C: 120 / 30 = 4

Como la mochila tiene capacidad 50, se toma el objeto A completo, el objeto B completo y una parte del objeto C.

La complejidad principal está en ordenar los objetos, por lo que el algoritmo tiene una complejidad aproximada de:

$$
O(n \log n)
$$

# 2. Backtracking: Resolución de Sudoku

## 2.1 Descripción del problema

El Sudoku es un tablero de 9x9 donde se deben completar las celdas vacías con números del 1 al 9, cumpliendo estas reglas:

- No se puede repetir un número en una misma fila.
- No se puede repetir un número en una misma columna.
- No se puede repetir un número en una subcuadrícula de 3x3.

## 2.2 Idea del algoritmo

El algoritmo utiliza **backtracking**:

1. Busca una celda vacía.
2. Prueba números del 1 al 9.
3. Verifica si el número cumple las reglas del Sudoku.
4. Si es válido, continúa con la siguiente celda.
5. Si más adelante no se puede continuar, retrocede y prueba otro número.

Es backtracking porque construye una solución paso a paso y descarta caminos que no llevan a una solución válida.


```python
def imprimir_tablero(tablero):
    """Imprime el tablero de Sudoku con un formato legible."""
    for fila in range(9):
        if fila % 3 == 0 and fila != 0:
            print("-" * 21)

        for columna in range(9):
            if columna % 3 == 0 and columna != 0:
                print("|", end=" ")

            print(tablero[fila][columna] if tablero[fila][columna] != 0 else ".", end=" ")

        print()


def buscar_celda_vacia(tablero):
    """Busca la primera celda vacía del tablero. Las celdas vacías se representan con 0."""
    for fila in range(9):
        for columna in range(9):
            if tablero[fila][columna] == 0:
                return fila, columna

    return None


def es_valido(tablero, numero, fila, columna):
    """Verifica si un número puede colocarse en una posición determinada."""

    # Validar fila
    if numero in tablero[fila]:
        return False

    # Validar columna
    for i in range(9):
        if tablero[i][columna] == numero:
            return False

    # Validar subcuadrícula 3x3
    inicio_fila = (fila // 3) * 3
    inicio_columna = (columna // 3) * 3

    for i in range(inicio_fila, inicio_fila + 3):
        for j in range(inicio_columna, inicio_columna + 3):
            if tablero[i][j] == numero:
                return False

    return True


def resolver_sudoku(tablero):
    """Resuelve el Sudoku utilizando backtracking."""

    celda_vacia = buscar_celda_vacia(tablero)

    # Si no hay celdas vacías, el Sudoku está resuelto
    if not celda_vacia:
        return True

    fila, columna = celda_vacia

    # Probar números del 1 al 9
    for numero in range(1, 10):
        if es_valido(tablero, numero, fila, columna):
            tablero[fila][columna] = numero

            # Intentar resolver el resto del tablero
            if resolver_sudoku(tablero):
                return True

            # Si no funcionó, se deshace la decisión
            tablero[fila][columna] = 0

    return False


# Sudoku de prueba
sudoku = [
    [5, 3, 0, 0, 7, 0, 0, 0, 0],
    [6, 0, 0, 1, 9, 5, 0, 0, 0],
    [0, 9, 8, 0, 0, 0, 0, 6, 0],

    [8, 0, 0, 0, 6, 0, 0, 0, 3],
    [4, 0, 0, 8, 0, 3, 0, 0, 1],
    [7, 0, 0, 0, 2, 0, 0, 0, 6],

    [0, 6, 0, 0, 0, 0, 2, 8, 0],
    [0, 0, 0, 4, 1, 9, 0, 0, 5],
    [0, 0, 0, 0, 8, 0, 0, 7, 9]
]

print("Sudoku inicial:")
imprimir_tablero(sudoku)

if resolver_sudoku(sudoku):
    print("\nSudoku resuelto:")
    imprimir_tablero(sudoku)
else:
    print("No se encontró solución.")
```

## 2.3 Análisis

El algoritmo prueba posibles valores en cada celda vacía y continúa solamente si la solución parcial es válida.

Cuando encuentra un conflicto, vuelve atrás y cambia la decisión anterior.  
Esa es la característica principal del **backtracking**.

En el peor caso, el algoritmo puede probar muchas combinaciones, por lo que su complejidad es alta. Sin embargo, las restricciones del Sudoku permiten descartar muchas opciones inválidas durante el proceso.

# 3. Divide y vencerás: Merge Sort

## 3.1 Descripción del problema

Merge Sort es un algoritmo de ordenamiento que permite ordenar una lista de elementos.

## 3.2 Idea del algoritmo

La técnica de **divide y vencerás** consiste en:

1. Dividir el problema en subproblemas más pequeños.
2. Resolver cada subproblema de forma recursiva.
3. Combinar las soluciones parciales.

En Merge Sort, la lista se divide en dos mitades hasta llegar a listas de un solo elemento. Luego, esas listas se combinan de forma ordenada.


```python
def merge_sort(lista):
    """Ordena una lista usando el algoritmo Merge Sort."""

    # Caso base: una lista vacía o de un elemento ya está ordenada
    if len(lista) <= 1:
        return lista

    mitad = len(lista) // 2

    # Dividir la lista en dos partes
    izquierda = lista[:mitad]
    derecha = lista[mitad:]

    # Ordenar cada parte de forma recursiva
    izquierda_ordenada = merge_sort(izquierda)
    derecha_ordenada = merge_sort(derecha)

    # Combinar ambas partes ordenadas
    return combinar(izquierda_ordenada, derecha_ordenada)


def combinar(izquierda, derecha):
    """Combina dos listas ordenadas en una sola lista ordenada."""

    resultado = []
    i = 0
    j = 0

    while i < len(izquierda) and j < len(derecha):
        if izquierda[i] <= derecha[j]:
            resultado.append(izquierda[i])
            i += 1
        else:
            resultado.append(derecha[j])
            j += 1

    # Agregar los elementos restantes
    resultado.extend(izquierda[i:])
    resultado.extend(derecha[j:])

    return resultado


# Lista de prueba
lista = [38, 27, 43, 3, 9, 82, 10]

print("Lista original:", lista)

lista_ordenada = merge_sort(lista)

print("Lista ordenada:", lista_ordenada)
```

## 3.3 Análisis

Merge Sort divide la lista en mitades hasta llegar a listas pequeñas.  
Luego combina esas partes de forma ordenada.

La complejidad de Merge Sort es:

$$
O(n \log n)
$$

Esto ocurre porque la lista se divide repetidamente en dos partes y, en cada nivel de división, se recorren los elementos para combinarlos.

# 4. Comparación de técnicas

| Técnica | Algoritmo utilizado | Estrategia principal | Tipo de problema |
|---|---|---|---|
| Voraz | Mochila fraccionaria | Elegir la mejor opción local en cada paso | Optimización |
| Backtracking | Sudoku | Probar soluciones y retroceder ante conflictos | Búsqueda de solución válida |
| Divide y vencerás | Merge Sort | Dividir, resolver y combinar | Ordenamiento |

## Conclusión

En esta actividad se implementaron tres técnicas diferentes de diseño algorítmico.

La **mochila fraccionaria** muestra cómo una estrategia voraz puede resolver un problema de optimización cuando se permite tomar fracciones de objetos.  
El **Sudoku** demuestra el uso de backtracking, ya que el algoritmo prueba valores posibles y retrocede cuando encuentra una contradicción.  
Finalmente, **Merge Sort** aplica divide y vencerás al separar una lista en partes más pequeñas, resolverlas y combinarlas ordenadamente.

Cada técnica tiene una forma distinta de abordar los problemas, por lo que es importante reconocer cuándo conviene utilizar cada una.
